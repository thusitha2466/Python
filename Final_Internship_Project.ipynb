```python
import pandas as pd

df = pd.read_csv("student_performance_dataset_1000_rows.csv")

print("Total Rows:", len(df))
print("Total Columns:", len(df.columns))

pd.set_option('display.max_rows', None)

print(df)
```

```python
print(df['Grade'].value_counts())
```

```python
print(df['Grade'].value_counts())
```

```python
print("Average Marks:", df['Average'].mean())
```

```python
top10 = df.sort_values(by='Average', ascending=False)
print(top10.head(10))
```

```python
import matplotlib.pyplot as plt

df['Grade'].value_counts().plot(kind='bar')

plt.title("Grade Distribution")
plt.xlabel("Grade")
plt.ylabel("Number of Students")
plt.show()
```

```python
import matplotlib.pyplot as plt

df['Pass_Fail'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Pass vs Fail Distribution")
plt.ylabel("")
plt.show()
```

```python
subjects = ['Tamil','English','Mathematics','Science']

print(df[subjects].mean())
```

```python
import matplotlib.pyplot as plt

df[subjects].mean().plot(kind='bar')

plt.title("Subject Wise Average Marks")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.show()
```
